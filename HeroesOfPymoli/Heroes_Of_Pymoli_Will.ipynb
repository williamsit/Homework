```python
#Dependencies
import os
import csv
import numpy as np
import pandas as pd
```

```python
#Read file as csv
heroes_file_loc = os.path.join("Resources", "purchase_data.csv")
heroes_file_reader = pd.read_csv(heroes_file_loc)
heroes_file_reader.head(2)
```

```python
#Rename "SN" column to "Screen name" for better understanding
heroes_file_renamed = heroes_file_reader.rename(columns={"SN":"Screen name"})
```

```python
#################### Player Count ###############################################
```

```python
#Player count
list_of_unique_screennames = heroes_file_renamed["Screen name"].unique()
count_of_screennames = len(list_of_unique_screennames)
print(count_of_screennames)

#Creating Dataframe
screennames = pd.DataFrame({"Total Players":[count_of_screennames]})
screennames
```

```python
#################### Purchasing Analysis (Total) #############################
```

```python
#Number of Unique Items
list_of_unique_item_name = heroes_file_renamed["Item ID"].unique()
count_of_unique_items = len(list_of_unique_item_name)

#Average Purchase Price
average_price = heroes_file_renamed["Price"].mean()
formatted_avg_price = "${:.2f}".format(average_price)

#Total Number of Purchases
num_purchases = heroes_file_renamed["Purchase ID"].count()

#Total Revenue
total_rev = heroes_file_renamed["Price"].sum()
formatted_total_rev = "${:.2f}".format(total_rev)

#Creating dataframe with combined information
purchasing_anal_total_df = pd.DataFrame({"Number of Unique Items":[count_of_unique_items],
                          "Average Purchase Price":[formatted_avg_price],
                          "Total Number of Purchases":[num_purchases],
                          "Total Revenue":[formatted_total_rev]})
purchasing_anal_total_df
```

```python
###########################Gender Demographics###########################
```

```python
#Percentage and Count of Male, Female, and Neither/Nondisclosed Players
```

```python
#Creating a groupby for Genders
grouped_gender_df = heroes_file_renamed.groupby(['Gender'])

#Total Count of Genders
total_gender = grouped_gender_df["Screen name"].nunique()

#Percentage of Genders
pct_gender = total_gender/count_of_screennames * 100
formatted_gender_female = "{:.2f}%".format(pct_gender[0])
formatted_gender_male = "{:.2f}%".format(pct_gender[1])
formatted_gender_neither = "{:.2f}%".format(pct_gender[2])
formatted_pct_gender = [formatted_gender_female, formatted_gender_male, formatted_gender_neither]

#Creating dataframe with combined information & sorting
gender_demo_df = pd.DataFrame({"Percentage of Players": formatted_pct_gender,
                              "Total Players": total_gender})
gender_demo_df_sorted = gender_demo_df.sort_values(["Total Players"], ascending = False)
gender_demo_df_sorted
```

```python
##################Purchasing Analysis (Gender)########################
```

```python
#Purchase Count
grouped_df_gender_purchase_count = grouped_gender_df["Purchase ID"].count()

#Average Purchase Price
grouped_df_gender_avg_purchase = grouped_gender_df["Price"].mean()
formatted_avg_female = "${:.2f}".format(grouped_df_gender_avg_purchase[0])
formatted_avg_male = "${:.2f}".format(grouped_df_gender_avg_purchase[1])
formatted_avg_neither = "${:.2f}".format(grouped_df_gender_avg_purchase[2])
formatted_grouped_df_gender_avg_purchase = [formatted_avg_female, formatted_avg_male, formatted_avg_neither]

#Total Purchase Value

grouped_df_gender_total_value = grouped_gender_df["Price"].sum()
formatted_total_female = "${:.2f}".format(grouped_df_gender_total_value[0])
formatted_total_male = "${:.2f}".format(grouped_df_gender_total_value[1])
formatted_total_neither = "${:.2f}".format(grouped_df_gender_total_value[2])
formatted_grouped_df_gender_total_value = [formatted_total_female, formatted_total_male, formatted_total_neither]

#Average Purchase Total per Person by Gender
grouped_df_gender_avg_per_person = grouped_df_gender_total_value/total_gender
formatted_avg_per_female = "${:.2f}".format(grouped_df_gender_avg_per_person[0])
formatted_avg_per_male = "${:.2f}".format(grouped_df_gender_avg_per_person[1])
formatted_avg_per_neither = "${:.2f}".format(grouped_df_gender_avg_per_person[2])
formatted_grouped_df_gender_avg_per_person = [formatted_avg_per_female, formatted_avg_per_male, formatted_avg_per_neither]

#Creating a dataframe with combined information & sorting
purchase_anal_gender_df = pd.DataFrame({"Purchase Count": grouped_df_gender_purchase_count,
                                       "Average Purchase Price": formatted_grouped_df_gender_avg_purchase,
                                       "Total Purchase Value": formatted_grouped_df_gender_total_value,
                                       "Average Purchase Total per Person": formatted_grouped_df_gender_avg_per_person})
purchase_anal_gender_df_sorted = purchase_anal_gender_df.sort_values(["Purchase Count"], ascending = False)
purchase_anal_gender_df_sorted.head(3)
```

```python
###########################Age Demographics#####################
```

```python
#Creating age group bins and new groupby
age_range = [0, 9.5, 14.5, 19.5, 24.5, 29.5, 34.5, 39.5, 44.5 , 150]
age_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40-44", "45+"]
heroes_file_renamed["Age Groups"] = pd.cut(heroes_file_renamed["Age"], age_range, labels = age_labels)
grouped_df_age_group = heroes_file_renamed.groupby("Age Groups")

#Total count per age
total_count_by_age_df = grouped_df_age_group["Screen name"].nunique()

#Percentage per age
pct_by_age_df = total_count_by_age/count_of_screennames * 100
formatted_pct_0 = "{:.2f}%".format(pct_by_age[0])
formatted_pct_1 = "{:.2f}%".format(pct_by_age[1])
formatted_pct_2 = "{:.2f}%".format(pct_by_age[2])
formatted_pct_3 = "{:.2f}%".format(pct_by_age[3])
formatted_pct_4 = "{:.2f}%".format(pct_by_age[4])
formatted_pct_5 = "{:.2f}%".format(pct_by_age[5])
formatted_pct_6 = "{:.2f}%".format(pct_by_age[6])
formatted_pct_7 = "{:.2f}%".format(pct_by_age[7])
formatted_pct_8 = "{:.2f}%".format(pct_by_age[8])
formatted_pct_by_age_df = [formatted_pct_0, formatted_pct_1, formatted_pct_2, formatted_pct_3, formatted_pct_4,
                          formatted_pct_5, formatted_pct_6, formatted_pct_7, formatted_pct_8]

#Creating dataframe with combined information
age_demographics_df = pd.DataFrame({"Total Count per Age":total_count_by_age_df,
                                   "Percentage of Players per Age":formatted_pct_by_age_df})
age_demographics_df.head(10)
```

```python
#########################Purchasing Analysis (Age)#############################
```

```python
#Purchase Count
grouped_df_age_purchase_count = grouped_df_age_group["Purchase ID"].count()

#Average Purchase Price
grouped_df_age_avg_purchase_price = grouped_df_age_group["Price"].mean()
formatted_avg_age_0 = "${:.2f}".format(grouped_df_age_avg_purchase_price[0])
formatted_avg_age_1 = "${:.2f}".format(grouped_df_age_avg_purchase_price[1])
formatted_avg_age_2 = "${:.2f}".format(grouped_df_age_avg_purchase_price[2])
formatted_avg_age_3 = "${:.2f}".format(grouped_df_age_avg_purchase_price[3])
formatted_avg_age_4 = "${:.2f}".format(grouped_df_age_avg_purchase_price[4])
formatted_avg_age_5 = "${:.2f}".format(grouped_df_age_avg_purchase_price[5])
formatted_avg_age_6 = "${:.2f}".format(grouped_df_age_avg_purchase_price[6])
formatted_avg_age_7 = "${:.2f}".format(grouped_df_age_avg_purchase_price[7])
formatted_avg_age_8 = "${:.2f}".format(grouped_df_age_avg_purchase_price[8])
formatted_grouped_df_age_avg_purchase_price = [formatted_avg_age_0, formatted_avg_age_1, formatted_avg_age_2, formatted_avg_age_3,
                                              formatted_avg_age_4, formatted_avg_age_5, formatted_avg_age_6, formatted_avg_age_7,
                                              formatted_avg_age_8]

#Total Purchase Value
grouped_df_age_total_purchase_value = grouped_df_age_group["Price"].sum()
formatted_total_age_0 = "${:.2f}".format(grouped_df_age_total_purchase_value[0])
formatted_total_age_1 = "${:.2f}".format(grouped_df_age_total_purchase_value[1])
formatted_total_age_2 = "${:.2f}".format(grouped_df_age_total_purchase_value[2])
formatted_total_age_3 = "${:.2f}".format(grouped_df_age_total_purchase_value[3])
formatted_total_age_4 = "${:.2f}".format(grouped_df_age_total_purchase_value[4])
formatted_total_age_5 = "${:.2f}".format(grouped_df_age_total_purchase_value[5])
formatted_total_age_6 = "${:.2f}".format(grouped_df_age_total_purchase_value[6])
formatted_total_age_7 = "${:.2f}".format(grouped_df_age_total_purchase_value[7])
formatted_total_age_8 = "${:.2f}".format(grouped_df_age_total_purchase_value[8])
formatted_grouped_df_age_total_purchase_value = [formatted_total_age_0, formatted_total_age_1, formatted_total_age_2, formatted_total_age_3,
                                                 formatted_total_age_4, formatted_total_age_5, formatted_total_age_6, formatted_total_age_7,
                                                 formatted_total_age_8]

#Average Purchase Total per Person by Age Group
grouped_df_age_avg_per_person = grouped_df_age_total_purchase_value/total_count_by_age_df
formatted_avg_per_age_0 = "${:.2f}".format(grouped_df_age_avg_per_person[0])
formatted_avg_per_age_1 = "${:.2f}".format(grouped_df_age_avg_per_person[1])
formatted_avg_per_age_2 = "${:.2f}".format(grouped_df_age_avg_per_person[2])
formatted_avg_per_age_3 = "${:.2f}".format(grouped_df_age_avg_per_person[3])
formatted_avg_per_age_4 = "${:.2f}".format(grouped_df_age_avg_per_person[4])
formatted_avg_per_age_5 = "${:.2f}".format(grouped_df_age_avg_per_person[5])
formatted_avg_per_age_6 = "${:.2f}".format(grouped_df_age_avg_per_person[6])
formatted_avg_per_age_7 = "${:.2f}".format(grouped_df_age_avg_per_person[7])
formatted_avg_per_age_8 = "${:.2f}".format(grouped_df_age_avg_per_person[8])
formatted_grouped_df_age_avg_per_person = [formatted_avg_per_age_0, formatted_avg_per_age_1, formatted_avg_per_age_2,
                                          formatted_avg_per_age_3, formatted_avg_per_age_4, formatted_avg_per_age_5,
                                          formatted_avg_per_age_6, formatted_avg_per_age_7, formatted_avg_per_age_8]

#Creating a dataframe with combined information
age_anal_grouped_df = pd.DataFrame({"Purchase Count":grouped_df_age_purchase_count,
                                   "Average Purchase Price":formatted_grouped_df_age_avg_purchase_price,
                                   "Total Purchase Value":formatted_grouped_df_age_total_purchase_value,
                                   "Average Purchase Total per Person by Age Group":formatted_grouped_df_age_avg_per_person})
age_anal_grouped_df.head(10)
```

```python
########################Top Spenders##############################
```

```python
#Groupby
grouped_df_spender = heroes_file_renamed.groupby("Screen name")

#Purchase Count
grouped_df_sn_purchase_count = grouped_df_spender["Purchase ID"].count()

#Average Purchase Price
grouped_df_sn_avg_purchase_price = grouped_df_spender["Price"].mean()

#Total Purchase Value
grouped_df_sn_total_purchase_value = grouped_df_spender["Price"].sum()

#Creating a dataframe with combined information & sorting & styling format
top_spenders_grouped_df = pd.DataFrame({"Purchase Count":grouped_df_sn_purchase_count,
                                       "Average Purchase Price":grouped_df_sn_avg_purchase_price,
                                       "Total Purchase Value":grouped_df_sn_total_purchase_value})
top_spenders_grouped_df_sorted = top_spenders_grouped_df.sort_values(["Total Purchase Value"], ascending=False).head(5)
top_spenders_grouped_df_sorted.style.format({"Average Purchase Price":"${:.2f}", "Total Purchase Value":"${:.2f}"})
```

```python
#######################Most Popular Items#########################
```

```python
#Groupby
grouped_df_pop_items = heroes_file_renamed.groupby(["Item ID", "Item Name"])

#Purchase Count
grouped_df_pop_purchase_count = grouped_df_pop_items["Purchase ID"].count()

#Total Purchase Value
grouped_df_pop_total_purchase_value = grouped_df_pop_items["Price"].sum()

#Item Price
grouped_df_pop_item_price = grouped_df_pop_total_purchase_value/grouped_df_pop_purchase_count

#Creating a dataframe with combined information & sorting
most_popular_items_summary_df = pd.DataFrame({"Purchase Count":grouped_df_pop_purchase_count,
                                              "Item Price":grouped_df_pop_item_price,
                                              "Total Purchase Value":grouped_df_pop_total_purchase_value})
most_popular_items_summary_df_sorted = most_popular_items_summary_df.sort_values(["Purchase Count"], ascending=False).head(5)
most_popular_items_summary_df_sorted.style.format({"Item Price":"${:.2f}", "Total Purchase Value":"${:.2f}"})
```

```python
########################Most Profitable Items########################
```

```python
#Groupby
grouped_df_prof_items = heroes_file_renamed.groupby(["Item ID", "Item Name"])

#Purchase Count
grouped_df_prof_purchase_count = grouped_df_prof_items["Purchase ID"].count()

#Total Purchase Value
grouped_df_prof_total_purchase_value = grouped_df_prof_items["Price"].sum()

#Item Price
grouped_df_prof_item_price = grouped_df_prof_total_purchase_value/grouped_df_prof_purchase_count

#Creating a dataframe with combined information & sorting
most_profitable_items_summary_df = pd.DataFrame({"Purchase Count":grouped_df_prof_purchase_count,
                                                "Item Price":grouped_df_prof_item_price,
                                                "Total Purchase Value":grouped_df_prof_total_purchase_value})
most_profitable_items_summary_df_sorted = most_profitable_items_summary_df.sort_values(["Total Purchase Value"], ascending = False).head(5)
most_profitable_items_summary_df_sorted.style.format({"Item Price":"${:.2f}", "Total Purchase Value":"${:.2f}"})

```
