```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from pprint import pprint
import os

# Import API key
from config import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = os.path.join("Resources", "output_data", "cities.csv")

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    city = city.replace(" ", "%20")
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python
#Creating url with API key
weather_key = api_key
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_key
```

```python
#Creating empty lists to extract data from json
city_name = []
city_cloudiness = []
city_country = []
city_date = []
city_humidity = []
city_lat = []
city_lng = []
city_max_temp = []
city_wind_speed = []

#Creating record counter and set count
record_count = 1
set_count = 1

#Printing log intro
print("Beginning Data Retrieval")
print("----------------------------------")

#Iterating through cities list and appending data to empty lists
for city in cities:
    try:
        url_response = requests.get(f"{url}&q={city}")
        url_json = url_response.json()
        city_title = url_json["name"]
        city_name.append(city_title)
        city_cloudiness.append(url_json["clouds"]["all"])
        city_country.append(url_json["sys"]["country"])
        city_date.append(url_json["dt"])
        city_humidity.append(url_json["main"]["humidity"])
        city_lat.append(url_json["coord"]["lat"])
        city_lng.append(url_json["coord"]["lon"])
        city_max_temp.append(url_json["main"]["temp_max"])
        city_wind_speed.append(url_json["wind"]["speed"])
        print(f"Processing Record {record_count} of Set {set_count} | {city_title}")
        
        if record_count < 50:
            record_count += 1
        elif record_count == 50:
            record_count = 1
            set_count += 1
        
        time.sleep(1.001)
    
    except KeyError:
        print("City not found. Skipping...")
    continue

print("----------------------------------")
print("Data Retrieval Complete")
print("----------------------------------")
```

```python
#Creating dictionary with lists
weather_dict = {
    "City": city_name,
    "Cloudiness": city_cloudiness,
    "Country": city_country,
    "Date": city_date,
    "Humidity": city_humidity,
    "Lat": city_lat,
    "Lng": city_lng,
    "Max Temp": city_max_temp,
    "Wind Speed": city_wind_speed
}

#Creating dataframe
weather_df = pd.DataFrame(weather_dict)

#Saving dataframe to csv
weather_df.to_csv(output_data_file)
```
